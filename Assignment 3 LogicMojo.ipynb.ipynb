```python
import seaborn as sns
import pandas as pd
import numpy as np

def prepare_titanic_features() -> np.ndarray:
    try:
        # 1. Load dataset
        df = sns.load_dataset("titanic")
        if df is None or df.empty:
            raise ValueError("Dataset cannot be loaded or is empty")

        df = df.copy()

        # 2. Handle Missing Values
        df['age'].fillna(df['age'].median(), inplace=True)
        df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
        df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)

        # 3. Drop column safely
        if 'deck' in df.columns:
            df.drop(columns='deck', inplace=True)

        # 4. Encode categorical variables
        encoded_df = pd.get_dummies(
            df,
            columns=['sex', 'embarked'],
            drop_first=True,
            dtype=int
        )

        # 5. Select Features
        features = ['age', 'fare', 'sex_male', 'embarked_Q', 'embarked_S']

        missing_cols = [col for col in features if col not in encoded_df.columns]
        if missing_cols:
            raise KeyError(f"Missing expected columns after encoding: {missing_cols}")

        X = encoded_df[features].copy()

        # 6. Normalize
        num_cols = ['age', 'fare']
        data = X[num_cols].values

        mean = np.mean(data, axis=0)
        std = np.std(data, axis=0)

        if np.any(std == 0):
            raise ZeroDivisionError("Standard deviation is zero for one or more columns")

        scaled_data = (data - mean) / std
        X.loc[:, num_cols] = scaled_data

        # 7. Convert to NumPy
        X_array = X.values
        return X_array

    except KeyError as ke:
        raise KeyError(f"Missing required column: {ke}")

    except ValueError as ve:
        raise ValueError(f"Value error during preprocessing: {ve}")

    except ZeroDivisionError as ze:
        raise ValueError(f"Zero Division Error during preprocessing: {ze}")

    except Exception as e:
        raise Exception(f"Unexpected error occurred: {e}")


# ================= RUN =================

print(f"{'★' * 45} TEST RESULTS {'★' * 44}")
print("=" * 149)

try:
    X_final = prepare_titanic_features()
    print("Final Feature Matrix Shape:", X_final.shape)
    print(X_final[:5,:])
except Exception as e:
    print("Error:", e)
print("=" * 149)
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

def indentify_correlation_feature():

    try:
        iris = sns.load_dataset("iris")
        if iris is None or iris.empty:
            raise ValueError("Dataset cannot be loaded or is empty")
        corr_matrix = iris.corr(numeric_only = True).abs()
        mask = np.triu(np.ones(corr_matrix.shape), k = 1)
        upper_triangle = corr_matrix.where(mask.astype(bool))

        to_drop = [col for col in upper_triangle.columns if any(upper_triangle[col] > 0.9)]

        print("Highly correlated feature:",to_drop)
        print("\nBefore:", iris.shape)
        
        iris_reduced = iris.drop(columns = to_drop)
        
        print("After :", iris_reduced.shape)
        print("\nRemaining columns:", iris_reduced.columns.tolist())

    except ValueError as ve:
        raise ValueError(f"Value error during preprocessing: {ve}")
    except Exception as e:
        raise Exception(f"Unexpected error occured: {e}")

# ================= RUN =================

print(f"{'★' * 45} TEST RESULTS {'★' * 44}")
print("=" * 149)

try:
    indentify_correlation_feature()
except Exception as e:
    print("Error:", e)
print("=" * 149) 
print(f"{"★" * 45} EXPLANATION {"★" * 45}")
print("Explain: Impact on overfitting and model performance")
print(
    " ==>> Avoid multicollinearity - Highly correlated feature = duplicate information. Learn noise instead of patterns, reduce redundancy, imporoves accuracy, faster training and leads to better performance ")
```

```python
import seaborn as sns

def highest_survival_age_group():
    try:
        df = sns.load_dataset("titanic")
        if df is None or df.empty:
            raise ValueError("Dataset is not loaded or is empty")
        df['age'].fillna(df['age'].median(), inplace=True)

        df['age_group'] = df['age'].apply(
            lambda x:"Child" if x < 18 else("Adult" if x < 60 else "Senior"))

        survival_rate_age_group = df.groupby('age_group')['survived'].mean()
        print("Survival Rate for each age group:\n",survival_rate_age_group)
        
    except ValueError as ve:
        raise ValueError(f"Value error during preprocessing: {ve}")
    except Exception as e:
        raise Exception(f"Unexpected error occured: {e}")

# ================= RUN =================

print(f"{'★' * 45} TEST RESULTS {'★' * 44}")
print("=" * 149)

try:
    highest_survival_age_group()
except Exception as e:
    print("Error:", e)
print("=" * 149) 
print(f"{"★" * 45} EXPLANATION {"★" * 44}")
print("Interpret: Which segment has highest survival likelihood")
print(
    " ==>> Child has high survival rate, Adult has mixed survival chances, Senior has less survival chances")
    
```

```python
import numpy as np

def error_prediction():

    try:
        rand_error = np.random.normal(loc=0,scale=1,size=1000)
        if rand_error is None:
            raise ValueError("Numpy array is not loaded or empty")
        mean = np.mean(rand_error)
        std = np.std(rand_error)

        lower_bound = mean - 2 * std
        upper_bound = mean + 2 * std

        outlier = rand_error[(rand_error < lower_bound) | (rand_error > upper_bound)]
        print("Count of Outliers:", len(outlier))

    except ValueError as ve:
        raise ValueError(f"Value error during preprocessing: {ve}")
    except Exception as e:
        raise Exception(f"Unexpected error occured: {e}")

# ================= RUN =================

print(f"{'★' * 45} TEST RESULTS {'★' * 44}")
print("=" * 149)

try:
    error_prediction()
except Exception as e:
    print("Error:", e)
print("=" * 149) 
print(f"{"★" * 45} EXPLANATION {"★" * 44}")
print("Explain: What high variance means in model performance")
print(
    " ==>> High variance means large predictions are spread out, model is unstable. Low variance, errors close to mean, model is stable")
    
```

```python
import seaborn as sns

def highest_revenue_segment():
    try:
        df = sns.load_dataset("tips")
        if df is None or df.empty:
            raise ValueError("Dataset is not loaded or is empty")
        df["tip_percentage"] = (df['tip'] / df['total_bill']) * 100

        grouped = df.groupby(['day','time']).agg({
            'total_bill': 'sum',
            'tip': 'sum',
            'tip_percentage': 'mean'
        }).reset_index()

        highest_revenue = grouped.loc[grouped['total_bill'].idxmax()]

        print("Highest revenue segment:\n",highest_revenue)

    except ValueError as ve:
        raise ValueError(f"Value error during preprocessing: {ve}")
    except Exception as e:
        raise Exception(f"Unexpected error occured: {e}")

# ================= RUN =================

print(f"{'★' * 45} TEST RESULTS {'★' * 44}")
print("=" * 149)

try:
    highest_revenue_segment()
except Exception as e:
    print("Error:", e)
print("=" * 149) 
print(f"{"★" * 45} EXPLANATION {"★" * 44}")
print("Provide: Recommendation for business strategy How ML model can use this insight")
print(
    " ==>> By analyzing tip_percentage and group by day and time, weekend dinners offer High revenue")
print(
    " ==>> This insight help ML models to learn and give personalized recommendations and optimizing business operations like increasing staffing,"
    "provide discount on weekdays, offer premium items to high bill cutomers")
    
```
