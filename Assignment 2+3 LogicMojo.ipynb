```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]

final = {}

def segment_customer(customer):
    value = customer["spend"]/customer["income"]
    if value > 0.6:
        return "High Value"
    elif value >= 0.3 and value <= 0.6:
        return "Medium Value"
    else:
        return "Low Value"

for customer in customers:
    segment = segment_customer(customer)
    final[customer["id"]] = segment
print(final)
```

```python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
cleaned_list = []
def clean_text(text):
    text =  re.sub("[^a-zA-Z0-9\s]","",text)
    text = text.lower()
    text = re.sub(r"\s+"," ",text)
    return text
for text in texts:
    cleaned_list = list(map(clean_text,texts))
print(cleaned_list)
```

```python
transactions = [
    {"id": 101, "amount": 5000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]

def is_fraud(transaction):
    if transaction["amount"] > 100000 or transaction["location"] == "Unknown":
        transaction["status"] = "fraud"
        return transaction
    else:
        transaction["status"] = "safe"
    return transaction
updated = list(map(is_fraud, transactions))

print(list(filter(lambda l:l["status"] == "fraud", updated)))
    
```

```python
def detect_increasing_sequence(data):

    if not data:
        return []

    current_seq = [data[0]]
    longest_seq = []
    for i in range(1,len(data)):

        if data[i] > data[i-1]:
            current_seq.append(data[i])
        else:
            if len(current_seq) > len(longest_seq):
                longest_seq = current_seq
            current_seq = [data[i]]
    if len(current_seq) > len(longest_seq):
        longest_seq = current_seq

    return longest_seq

data = [10, 20, 30, 25, 40, 50]
print(detect_increasing_sequence(data))
    
```

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

def recommend(user):
    target_items = set(user_purchases[user])
    recommendations = set()

    for other_user, items in user_purchases.items():
        if other_user == user:
            continue

        other_items = set(items)

        if target_items & other_items:
            recommendations |= other_items

    recommendations -= target_items
    return list(recommendations)

print(recommend("user2"))
        
```

```python
def decision_tree(score):
    if score > 80:
        return "Approved"
    elif score >= 50 and score <=80:
        return decision_tree(score+10)
    elif score < 50:
        return "Rejected"

print(decision_tree(50))
```
